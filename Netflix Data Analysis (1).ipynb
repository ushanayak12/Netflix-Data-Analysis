# Visualization Of different columns
1. Content Type Distribution (TV Shows vs Movies)

```python
data['type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

2. Count of Content Over the Years

```python
data['year_added'] = pd.DatetimeIndex(data['date_added']).year
data['year_added'].value_counts().sort_index().plot(kind='line')
```

3.  Genre (listed_in) / type Insights

```python
from collections import Counter
genres = data['listed_in'].str.split(', ')
genre_list = [item for sublist in genres.dropna() for item in sublist]
pd.Series(genre_list).value_counts().head(10).plot(kind='bar')
```

4. Movie Duration Distribution

```python
import matplotlib.pyplot as plt

# Filter movies only and ensure it's a copy
movie_data = data[data['type'] == 'Movie'].copy()

# Remove rows where duration doesn't contain "min"
movie_data = movie_data[movie_data['duration'].str.contains('min', na=False)]

# Clean and convert to float
movie_data['Minutes'] = movie_data['duration'].str.replace(' min', '', regex=False).astype(float)

# Plot
movie_data['Minutes'].plot(kind='hist', bins=30, color='teal', edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

# HeatMap for the missing values.
--> The netflix dataset has many missing values in key columns like director, cast, and country.
--> Visualize this helps you understand data quality before analysis ot model building.



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.heatmap(data.isnull(), cbar = False, cmap = 'viridis', yticklabels = False)
plt.title('Missing Values Heatmap - Netflix Dataset')
plt.show()
```

# Correlation Heatmap (numeric features)
--> This shows the correlation between numeric columns(eg. mintues, year_added etc).

```python
data.head()
```

```python
# ensure numeric columns like minutes and year_added exist
data['year_added'] = pd.to_datetime(data['date_added'], errors = 'coerce').dt.year
data['minutes'] = data['duration'].str.extract('(\d+)').astype(float)

# selecr numeric columns
numeric_columns = data.select_dtypes(include = 'number')

# plot correlation heatmap
plt.figure(figsize = (8, 6))
sns.heatmap(numeric_columns.corr(), annot = True, cmap = 'coolwarm', linewidth = 0.5)
plt.title('Correlation Heatmap of netflix numeric features')
plt.show()
```
